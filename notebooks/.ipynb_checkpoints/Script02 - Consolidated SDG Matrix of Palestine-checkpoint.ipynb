# Consolidated SDG Matrix of Palestine

Read SDG data provided in multi-dimensional tables

```python
import pandas as pd
import os 

dir_path = os.path.dirname(os.path.realpath('__file__'))
print(dir_path)

global_indicator_dir = r'../Global Indicator Framework/'


print('global indicator framework dir: ' + global_indicator_dir)


output_dir = r'../SDG-Matrix/Output/'
print('sdg matix outputs dir: ' + output_dir)

```

```python
indicator_list = pd.read_excel(output_dir + 'consolidated_mdt_REV4.xlsx')
indicator_list.head(5)

```

```python
print(indicator_list.shape)
print(indicator_list.columns)
```

```python
pcbs_list = indicator_list[['INDICATOR_CODE', 'INDICATOR_DESC_EN', 'SUBINDICATOR_CODE','SUBINDICATOR_DESC_AR', 'SUBINDICATOR_DESC_EN', 'UNIT_MEASURE', 'UNIT_MULT', 'BASE_PERIOD', ]].drop_duplicates()
pcbs_list['side'] = 'pcbs'
pcbs_list.head(5)
```

```python
print(pcbs_list.shape)
print(pcbs_list.columns)
```

```python
pcbs_list.to_excel(output_dir + 'pcsb_series.xlsx',index=False, encoding='UTF-8')
```

## Read global series catalouge

```python
global_series_catalogue = pd.read_excel(global_indicator_dir + 'global_series_catalogue.xlsx')
indicator_list.head(5)
```

Create a side-by side view:

```python
x = pd.merge(global_series_catalogue, pcbs_list, how = 'outer', 
         left_on = ['IndicatorCode', 'side'],
         right_on = ['INDICATOR_CODE', 'side']
         )
x.head(5)
```

```python
x['common_IndicatorCode'] = x['IndicatorCode']
x.common_IndicatorCode.fillna(x.INDICATOR_CODE, inplace=True)
x['common_IndicatorCode'].head(10)
```

```python
x.head(10)
```

```python
x.to_excel(output_dir + 'global_x_pcsb_series.xlsx',index=False, encoding='UTF-8')
```

# Data cleanup

At this stage, a manual proces of matching the data series available in PCSB with the data series available in the Global Indicator Database was carried out.  The main objective was to map `sub-indicators` definied in PCBS to `series` codes defined in the Global Indicator Database.

The resulting indicator catalogue is included in the excel file named `global_x_pcsb_series_edited2.xlsx`, and the consolidated data file is `consolidated_mdt_REV5.xlsx`
