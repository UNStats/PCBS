```python
import pandas as pd
import os 

dir_path = os.path.dirname(os.path.realpath('__file__'))
print(dir_path)

input_dir = r'../SDG-Matrix/Input/'
print('data inputs dir: ' + input_dir)

output_dir = r'../SDG-Matrix/Output/'
print('outputs dir: ' + output_dir)
```

```python
mdt2_df = pd.read_excel(output_dir + 'consolidated_mdt_REV3.xlsx', encoding='UTF-8', 
                              dtype={'GOAL': str,
                                    'TARGET': str})

mdt2_df.head(5)
```

```python
series_catalog = mdt2_df.drop(['REF_AREA',
                                'REF_AREA_DESC_AR',
                                'REF_AREA_DESC_EN',
                                'TIME_PERIOD',
                                'OBS_COMMENT',
                                'OBS_STATUS',
                                'OBS_VALUE',
                                'TIME_DETAIL',
                                'SOURCE_DETAIL',
                                'SOURCE_DETAIL_DESC_AR',
                                'SOURCE_DETAIL_DESC_EN'], axis=1).drop_duplicates()

series_catalog.head(10)
```
