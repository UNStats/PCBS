# Data cleanup

At this stage, a manual proces of matching the data series available in PCSB with the data series available in the Global Indicator Database was carried out.  The main objective was to map `sub-indicators` definied in PCBS to `series` codes defined in the Global Indicator Database.

The resulting indicator catalogue is included in the excel file named `global_x_pcsb_series_edited2.xlsx`, and the consolidated data file is `consolidated_mdt_REV5.xlsx`

## Open the indicator catalogue

```python
consolidated_indicator_list = pd.read_excel(output_dir + 'global_x_pcsb_series_edited2.xlsx')
consolidated_indicator_list.head(5)
```

## Open the consolidated data matrix

```python
consolidated_data_matrix = pd.read_excel(output_dir + 'consolidated_mdt_REV5.xlsx')
consolidated_data_matrix.head(5)
```

```python
print(consolidated_data_matrix.shape)
print(consolidated_data_matrix.columns)
```

```python
pcbs_list = consolidated_indicator_list.loc[consolidated_indicator_list['PCBS_Series'] == 1]
pcbs_list.head(4)
```

```python
print(pcbs_list.shape)
print(pcbs_list.columns)
```

```python
for row in pcbs_list.itertuples(index=True, name='Pandas'):
    x = consolidated_indicator_list.loc[(consolidated_indicator_list['Indicator'] == getattr(row, 'Indicator') )& (consolidated_indicator_list['Series'] == getattr(row, 'Series') )]
    x = pd.merge(x, 
                 consolidated_data_matrix[['INDICATOR_CODE', 'SERIES_CODE', 'AGE', 'EDUCATION_LEVEL', 'OCCUPATION', 'SEX', 
                                           'URBANISATION', 'INTERNET_SPEED', 'REF_AREA', 'TIME_PERIOD', 'OBS_VALUE', 
                                           'UNIT_MEASURE', 'UNIT_MULT', 'BASE_PERIOD', 'FOOTNOTE', 'TIME_DETAIL', 'SOURCE']],
                 how = 'left',
                 left_on = ['IndicatorCode', 'Series'],
                 right_on = ['INDICATOR_CODE', 'SERIES_CODE']
              )
    # x.to_excel(output_dir + 'mdt_' +  getattr(row, 'Indicator') + '_' +  getattr(row, 'Series')+'.xlsx',index=False, encoding='UTF-8')
    
    x_columns = ['Goal', 'GoalDesc', 
                 'Target', 'TargetDesc', 
                 'Indicator', 'IndicatorDesc', 'IndicatorCode', 
                 'Series', 'SeriesDescEN', 'SeriesDescAR', 
                 'AGE', 'EDUCATION_LEVEL', 'OCCUPATION', 'SEX', 'URBANISATION', 'INTERNET_SPEED', 
                 'REF_AREA', 'TIME_PERIOD', 
                 'OBS_VALUE', 'UNIT_MEASURE', 'UNIT_MULT', 'BASE_PERIOD', 
                 'FOOTNOTE', 'TIME_DETAIL', 'SOURCE', 
                 'GlobalDatabase', 'Comment']

    
    ##############################################################################
    #
    # An example of converting a Pandas dataframe to an xlsx file
    # with column formats using Pandas and XlsxWriter.
    #
    # Copyright 2013-2019, John McNamara, jmcnamara@cpan.org
    #
    
    sheet = getattr(row, 'Indicator') + '_' +  getattr(row, 'Series')

    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter(output_dir + 'mdt_' +  sheet +'.xlsx', engine='xlsxwriter')

    # Convert the dataframe to an XlsxWriter Excel object.
    x[x_columns].to_excel(writer, sheet_name=sheet, index=False, encoding='UTF-8')

    # Get the xlsxwriter workbook and worksheet objects.
    workbook  = writer.book
    worksheet = writer.sheets[sheet]

    # Add some cell formats.
    #format1 = workbook.add_format({'num_format': '#,##0.00'})
    #format2 = workbook.add_format({'num_format': '0%'})
    #cell_format = workbook.add_format()
    #cell_format.set_text_wrap()

    # Set the column width and format.
    #worksheet.set_column('A:AI', 24, cell_format)
    worksheet.set_column('A:AI', 24, None)

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```
